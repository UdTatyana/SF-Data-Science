```python
import pandas as pd
from IPython.display import display 
import numpy as np
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv', sep=',')
display(students_performance)
```

```python
students_performance.gender == 'female'
# Series с соответствием условию
```

```python
# выведем только True
students_performance.loc[students_performance.gender == 'female']
```

```python
students_performance.loc[students_performance.gender == 'female', ['gender', 'writing score']]
# вывели только нужные колонки
```

```python
mean_writing_score = students_performance['writing score'].mean()
# среднее по writing score
students_performance.loc[students_performance['writing score']>mean_writing_score]
```

```python
# объединим условия, сохраним в переменную
query = (students_performance['writing score']>mean_writing_score)&(students_performance.gender == 'female')
```

```python
# обратимся к DF, используя эту переменную
students_performance.loc[query].head()
```
