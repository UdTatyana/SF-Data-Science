## Задание 7.5
Для датасета с ценами квартир проведите кластеризацию DBSCAN с параметрами по умолчанию. Визуализируйте результаты. Сколько точек было размечено как выбросы? (Ответ: 2)


```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN

df = pd.read_csv('data/housing.csv')
clustering = DBSCAN().fit(df[['year', 'price']])
sns.scatterplot(x=df.year, y=df.price, c=clustering.labels_)
```

## Задание 7.6
В предыдущем задании алгоритм DBSCAN ошибочно определил точку у жёлтого кластера как выброс. Так произошло потому, что алгоритм запускался с параметрами по умолчанию.

Какие значения параметров eps и min_samples подойдут для верной кластеризации данных?

Отметьте все подходящие варианты ответов.

A eps = 0.5, min_samples = 5
B eps = 0.7, min_samples = 6
C eps = 0.7, min_samples = 5
D eps = 1.15, min_samples = 2

(Ответ: B, C)

```python
# A
clustering_A = DBSCAN(eps = 0.5, min_samples = 5).fit(df[['year', 'price']])
sns.scatterplot(x=df.year, y=df.price, c=clustering_A.labels_)
```

```python
# B
clustering_B = DBSCAN(eps = 0.7, min_samples = 6).fit(df[['year', 'price']])
sns.scatterplot(x=df.year, y=df.price, c=clustering_B.labels_)
```

```python
# C
clustering_C = DBSCAN(eps = 0.7, min_samples = 5).fit(df[['year', 'price']])
sns.scatterplot(x=df.year, y=df.price, c=clustering_C.labels_)
```

```python
# D
clustering_D = DBSCAN(eps = 1.15, min_samples = 2).fit(df[['year', 'price']])
sns.scatterplot(x=df.year, y=df.price, c=clustering_D.labels_)
```
