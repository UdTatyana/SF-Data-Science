Так как данные собирали все члены команды, нужно объединить датасэты по каждому фонду и удалить дубли

```python
import pandas as pd # импорт библиотек
from IPython.display import display

def get_df(df_1, df_2, df_3): # общая функция для обработки DF
   df = pd.concat([df_1, df_2, df_3]) # конкатенируем датасэты
   return df.drop_duplicates() # удаляем дубликаты строк

df_fond_abs_help = pd.read_csv('data/fond_abs_help.csv') # считываем первый файл
df_fond_abs_help_1 = pd.read_csv('data/fond_abs_help (1).csv') # считываем второй файл
df_fond_abs_help_2 = pd.read_csv('data/data_fomd_abs_help.csv') # считываем третий файл
fond_abs_help = get_df(df_fond_abs_help, df_fond_abs_help_1, df_fond_abs_help_2) # применяем к ним функцию, 
# результат помещаем в переменную

# точно так же применяем функцию get_df ко всем остальным файлам:

df_fond_agat = pd.read_csv('data/fond_agat.csv')
df_fond_agat_1 = pd.read_csv('data/fond_agat (1).csv')
df_fond_agat_2 = pd.read_csv('data/data_fomd_agat.csv')
fond_agat = get_df(df_fond_agat, df_fond_agat_1, df_fond_agat_2)

df_fond_bazis = pd.read_csv('data/fond_bazis.csv')
df_fond_bazis_1 = pd.read_csv('data/fond_bazis (1).csv')
df_fond_bazis_2 = pd.read_csv('data/data_fomd_bazis.csv')
fond_bazis = get_df(df_fond_bazis, df_fond_bazis_1, df_fond_bazis_2)

df_fond_dar = pd.read_csv('data/fond_dar.csv')
df_fond_dar_1 = pd.read_csv('data/fond_dar (1).csv')
df_fond_dar_2 = pd.read_csv('data/data_fond_dar.csv')
fond_dar = get_df(df_fond_dar, df_fond_dar_1, df_fond_dar_2)

df_fond_hamovniki = pd.read_csv('data/fond_hamovniki.csv')
df_fond_hamovniki_1 = pd.read_csv('data/fond_hamovniki (1).csv')
df_fond_hamovniki_2 = pd.read_csv('data/data_fond_hamovniki.csv')
fond_hamovniki = get_df(df_fond_hamovniki, df_fond_hamovniki_1, df_fond_hamovniki_2)

df_fond_ins = pd.read_csv('data/fond_ins.csv')
df_fond_ins_1 = pd.read_csv('data/fond_ins (1).csv')
df_fond_ins_2 = pd.read_csv('data/data_fomd_ins.csv')
fond_ins = get_df(df_fond_ins, df_fond_ins_1, df_fond_ins_2)

df_fond_m_prohor = pd.read_csv('data/fond_m_prohor.csv')
df_fond_m_prohor_1 = pd.read_csv('data/fond_m_prohor (1).csv')
df_fond_m_prohor_2 = pd.read_csv('data/data_fomd_m_prohor.csv')
fond_m_prohor = get_df(df_fond_m_prohor, df_fond_m_prohor_1, df_fond_m_prohor_2)

df_fond_melnichenko = pd.read_csv('data/fond_melnichenko.csv')
df_fond_melnichenko_1 = pd.read_csv('data/fond_melnichenko (1).csv')
df_fond_melnichenko_2 = pd.read_csv('data/data_fond_melnichenko.csv')
fond_melnichenko = get_df(df_fond_melnichenko, df_fond_melnichenko_1, df_fond_melnichenko_2)

df_fond_muslim_magomaev = pd.read_csv('data/fond_muslim_magomaev.csv')
df_fond_muslim_magomaev_1 = pd.read_csv('data/fond_muslim_magomaev (1).csv')
df_fond_muslim_magomaev_2 = pd.read_csv('data/data_fond_muslim_magomaev.csv')
fond_muslim_magomaev = get_df(df_fond_muslim_magomaev, df_fond_muslim_magomaev_1, df_fond_muslim_magomaev_2)

df_fond_noosphera = pd.read_csv('data/fond_noosphera.csv')
df_fond_noosphera_1 = pd.read_csv('data/fond_noosphera (1).csv')
df_fond_noosphera_2 = pd.read_csv('data/data_fond_noosphera.csv')
fond_noosphera = get_df(df_fond_noosphera, df_fond_noosphera_1, df_fond_noosphera_2)

df_fond_our_future = pd.read_csv('data/fond_our_future.csv')
df_fond_our_future_1 = pd.read_csv('data/fond_our_future (1).csv')
df_fond_our_future_2 = pd.read_csv('data/data_fomd_our_future.csv')
fond_our_future = get_df(df_fond_our_future, df_fond_our_future_1, df_fond_our_future_2)

df_fond_rybakov = pd.read_csv('data/fond_rybakov.csv')
df_fond_rybakov_1 = pd.read_csv('data/fond_rybakov (1).csv')
df_fond_rybakov_2 = pd.read_csv('data/data_fomd_rybakov.csv')
fond_rybakov = get_df(df_fond_rybakov, df_fond_rybakov_1, df_fond_rybakov_2)

df_fond_samfar = pd.read_csv('data/fond_samfar.csv')
df_fond_samfar_1 = pd.read_csv('data/fond_samfar (1).csv')
df_fond_samfar_2 = pd.read_csv('data/data_fomd_samfar.csv')
fond_samfar = get_df(df_fond_samfar, df_fond_samfar_1, df_fond_samfar_2)

df_fond_svet = pd.read_csv('data/fond_svet.csv')
df_fond_svet_1 = pd.read_csv('data/fond_svet (1).csv')
df_fond_svet_2 = pd.read_csv('data/data_fond_svet.csv')
fond_svet = get_df(df_fond_svet, df_fond_svet_1, df_fond_svet_2)

df_fond_volnoe_delo = pd.read_csv('data/fond_volnoe_delo.csv')
df_fond_volnoe_delo_1 = pd.read_csv('data/fond_volnoe_delo (1).csv')
df_fond_volnoe_delo_2 = pd.read_csv('data/data_fomd_volnoe_delo.csv')
fond_volnoe_delo = get_df(df_fond_volnoe_delo, df_fond_volnoe_delo_1, df_fond_volnoe_delo_2)

df_fondtimchenko = pd.read_csv('data/fondtimchenko.csv')
df_fondtimchenko_1 = pd.read_csv('data/fondtimchenko (1).csv')
df_fondtimchenko_2 = pd.read_csv('data/data_fondtimchenko.csv')
fondtimchenko = get_df(df_fondtimchenko, df_fondtimchenko_1, df_fondtimchenko_2)

df_potanin = pd.read_csv('data/potanin.csv')
df_potanin_1 = pd.read_csv('data/potanin (1).csv')
df_potanin_2 = pd.read_csv('data/data_potanin.csv')
potanin = get_df(df_potanin, df_potanin_1, df_potanin_2)

df_podarizhizn = pd.read_csv('data/podarizhizn (1).csv')
df_podarizhizn_1 = pd.read_csv('data/data_podarizhizn.csv')
podarizhizn = pd.concat([df_podarizhizn, df_podarizhizn_1])
podarizhizn = podarizhizn.drop_duplicates()
```

## Преобразование данных

Все датасэты имеют одинаковую структуру, посмотрим на неё.

```python
print(fond_samfar.shape) # проверим размерность DF
print(podarizhizn.shape) 
```

```python
print(fond_samfar.info()) # общая информация о столбцах
```

Удалим ненужные нам столбцы.

```python
fond_list = [fond_abs_help, fond_agat, fond_bazis, fond_dar, fond_hamovniki, fond_ins, fond_m_prohor, fond_melnichenko, \
fond_muslim_magomaev,  fond_noosphera,  fond_our_future, fond_rybakov, fond_samfar, fond_svet, fond_volnoe_delo,\
fondtimchenko, potanin, podarizhizn]

for fond in fond_list:
    fond.drop(['conversation_id', 'created_at', 'time', 'timezone', 'user_id', 'name', 'place', 'language', 'mentions',
       'urls', 'photos','cashtags', 'link', 'retweet', 'quote_url', 'video', 'thumbnail', 'near', 'geo', 'source', 'user_rt_id', 
       'user_rt','retweet_id', 'reply_to', 'retweet_date', 'translate', 'trans_src','trans_dest'],
       axis=1,
       inplace=True
    )
    
```

```python
display(fond_abs_help.head(1)) # проверим: выведем первую строку с названием столбцов
```

В структуре не хватает столбца-идентификатора фонда, и при конкатенации датасэтов в общий строки просто сольются без указания, к какому именно фонду они относятся. Поэтому в каждый датасэт добавим столбец 'fond_name'. Для удобства работы заполним этот столбец не полным наименоваем фонда, а только его названием или фамилией, фигугрирующей в официальном названии.

```python
fond_abs_help['fond_name'] = '"Абсолют-помощь"' # добавим новый признак - название фонда
fond_agat['fond_name'] = '"Агат"'
fond_bazis['fond_name'] = '"Базис"'
fond_dar['fond_name'] = '"Дар"'
fond_hamovniki['fond_name'] = '"Хамовники"'
fond_ins['fond_name'] = '"Искусство, наука и спорт"'
fond_m_prohor['fond_name'] = 'Фонд Прохорова'
fond_melnichenko['fond_name'] = 'Фонд Мельниченко'
fond_muslim_magomaev['fond_name'] = 'Фонд Муслима Магомаева'
fond_noosphera['fond_name'] = '"Ноосфера"'
fond_our_future['fond_name'] = '"Наше будущее"'
fond_rybakov['fond_name'] = 'Рыбаков фонд'
fond_samfar['fond_name'] = '"Сафмар"'
fond_svet['fond_name'] = '"Свет"'
fond_volnoe_delo['fond_name'] = '"Вольное дело"'
fondtimchenko['fond_name'] = 'Фонд Тимченко'
potanin['fond_name'] = 'Фонд Потанина'
podarizhizn['fond_name'] = '"Подари жизнь"'

display(fond_abs_help.head()) # посмотрим, как теперь выглядит датасэт
```

Теперь можно объединить датасэты в общий.

```python
data = pd.concat([fond_abs_help, fond_agat, fond_bazis, fond_dar, fond_hamovniki, fond_ins, fond_m_prohor, fond_melnichenko, \
fond_muslim_magomaev,  fond_noosphera,  fond_our_future, fond_rybakov, fond_samfar, fond_svet, fond_volnoe_delo,\
fondtimchenko, potanin, podarizhizn])

print(data.shape) # размерность нового DF
print(data.info()) # общая информация о столбцах
```

Переведём дату в формат Datetime

```python
data['date'] = pd.to_datetime(data['date'])
data.info() # проверим
```

## Предварительный анализ данных

Посмотрим на распределение общего количества твитов для кождого фонда. 

```python
import plotly # импорт библиотеки
import plotly.express as px # импорт модуля express

fig = px.histogram( # строим гисторгамму
    data_frame=data, # датасет
    x='fond_name', # признак по оси абсцисс
    title='График 1. Общее количество собранных сообщений в Twitter для каждого фонда', # название графика
)
fig.show()

fig.write_html("graphs/histogram.html") # сохранение в html
```

Количество твитов варьируется от 40 (фонд "Базис") до 508 ("Фонд Тимченко"). Реже всего упоминались фонд "Базис", фонд "Агат" и фонд "Сафмар", чаще всех взятых в выборку - "Фонд Тимченко" и фонд "Вольное дело". Большинство фондов упоминали в сообщениях в среднем от 100 до 150 раз.

Посмотрим на даты упоминаний.

```python
import plotly
import plotly.express as px

fig = px.line(
    data_frame=data,
    x='date',
    y='fond_name',
    color='fond_name',
    title='График 2. Временной интервал упоминаний для каждого фонда'
)
fig.show()

fig.write_html("graphs/histogram.html") # сохранение в html
```

Самый короткий перион зафиксированных упоминаний - о фонде "Свет" - с октября 2021 по июль 2022 года (115 твитов).
На графике видны 2 группы массовых твитов: с января 2011 и с декабря 2014 годов.

Посмотрим на частоту упоминаний фондов в месяц.
Возьмем только те графики, в отношении которых собрано больше всего информации (с ними планируется дальнейшая работа).

```python
df_month = pd.DataFrame() # создаем новый DataFrame
df_month['fond_name'] = data['fond_name'] # в нем будет столбец с фондами
df_month['date'] = data['date'] # и признак даты

print(df_month.info())
```

```python
import plotly
import plotly.express as px

df_month = pd.DataFrame() # создаем новый DataFrame
df_month['fond_name'] = data['fond_name'] # в нем будет столбец с фондами
df_month['date'] = data['date'] # и признак даты
df_month['date'] = pd.to_datetime(df_month['date'])
df_month['year-month'] = data['date'].apply(lambda x: x.strftime('%Y-%B')) # даты превращаем 
# в строку с упоминанием года и месяца
df_month['year-month'] = pd.to_datetime(df_month['year-month']) # переведем ее в datetime, чтобы откинуть дни

mask = (df_month['fond_name'] == '"Вольное дело"')|(df_month['fond_name'] == 'Фонд Тимченко')|(df_month['fond_name'] == '"Подари жизнь"')
bar_data = df_month[mask]


fig = px.bar(
    data_frame=bar_data,
    x='year-month',
    y='fond_name',
    color='fond_name',
    title='График 3. Количество упоминаний в месяц'
)
fig.show()
```
